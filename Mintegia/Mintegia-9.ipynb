```python
def histograma(fizena,kodif="utf8"):
    f = open(fizena,encoding=kodif)
    h = {}
    for ilara in f :
        for hitza in ilara.split():
            if hitza in h :
                h[hitza] += 1
            else :
                h[hitza] = 1
    f.close()   
    return h

histograma("berriak_utf8.txt")
```
