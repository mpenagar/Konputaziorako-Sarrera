## 1
Idatzi funtzio bat testu fitxategi	batean	dagoen hitzen	histograma	(hitz bakoitzaren
agerpen kopurua)	lortzeko


```python
def histograma(fizena,kodif="utf8"):
    f = open(fizena,encoding=kodif)
    h = {}
    for ilara in f :
        for hitza in ilara.split():
            if hitza in h :
                h[hitza] += 1
            else :
                h[hitza] = 1
    f.close()   
    return h

h = histograma("berriak_utf8.txt")
```

## 3
Programa bat idatzi, Python programazio lengoaia erabiliz, testu fitxategi baten
hitzen alderantzizko histograma adierazteko hiztegi bat erabiltzen duena. Giltza
bakoitza zenbaki oso bat izango da (n) eta berari dagokion balioa fitxategian n aldiz
agertu diren hitzez osoturiko lista.


```python
def amargotsih(h):
    h2 = {}
    for hitza,agerpen in h.items():
        if agerpen in h2 :
            h2[agerpen].append(hitza)
        else :
            h2[agerpen] = [hitza]
    return h2

h2 = amargotsih(h)
```

```python
def f3(fizena,kodif="utf8"):
    return amargotsih(histograma(fizena,kodif))

f3("berriak_utf8.txt")
```

Funtzio bat diseinatu irakasgai batean matrikulatutako ikasleei buruzko informazioa
gordetzeko python-eko datu egitura bat erabiltzen duena.

```python
def f4(fizena,kodif="utf8"):
    f = open(fizena,encoding=kodif)
    h = {}
    for ilara in f :
        z = ilara.split()
        # Zerrenda bat ezin da gako bezala erabili
        # g = z[:3]
        # testu bilaka dezakegu
        #g = " ".join(z[:3])
        g = tuple(z[:3])
        b = z[3:]
        h[g] = b
    f.close()   
    return h    

f4("ikasleak.txt")
```
