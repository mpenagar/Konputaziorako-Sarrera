### 1
Idatzi funtzio bat testu fitxategi batean dagoen hitzen histograma (hitz bakoitzaren agerpen kopurua) lortzeko.

```python
# ZERRENDEKIN
def histograma(bideizena, kodif='utf8'):
    f = open(bideizena,encoding=kodif)
    zhitz = []
    zkont = []
    for ilara in f :
        for hitza in ilara.split():
            if hitza in zhitz :
                pos = zhitz.index(hitza)
                zkont[pos] += 1
            else :
                zhitz.append(hitza)
                zkont.append(1)
    f.close()
    return zhitz,zkont
    
histograma('berriak_utf8.txt')
```

```python
# HIZTEGIEKIN
def histograma(bideizena, kodif='utf8'):
    f = open(bideizena,encoding=kodif)
    h = {}
    for ilara in f :
        for hitza in ilara.split():
            if hitza in h :
                h[hitza] = h[hitza] + 1
            else :
                h[hitza] = 1
    f.close()
    return h
    
histograma('berriak_utf8.txt')
```

```python
# HIZTEGIEKIN 2
def histograma(bideizena, kodif='utf8'):
    f = open(bideizena,encoding=kodif)
    h = {}
    for ilara in f :
        for hitza in ilara.split():
            h[hitza] = h.get(hitza,0) + 1
    f.close()
    return h
    
histograma('berriak_utf8.txt')
```
